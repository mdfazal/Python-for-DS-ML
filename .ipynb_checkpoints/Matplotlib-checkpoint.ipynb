# Matplotlib Concept Lecture 

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')

```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[1].plot(x,y)
axes[1].set_title('Second Plot')
axes[0].plot(y,x)
axes[0].set_title('First Plot')
```

# Figure Size and DPI

```python
fig = plt.plot(figsize=(8,2))

ax = fig.ax([0,0,1,1])
#ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X SQUARED')
ax.plot(x,x**3, label='X CUBED')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X SQUARED')
ax.plot(x,x**3, label='X CUBED')

ax.legend(loc=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X SQUARED')
ax.plot(x,x**3, label='X CUBED')

ax.legend(loc=(0.1,0.1))
```

# Plot Appearance

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green')
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00')
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green',linewidth=3, alpha=0.5)
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green',lw=3, alpha=0.5,linestyle='--')
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green',lw=3, alpha=0.5,linestyle='-.')
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green',lw=3, alpha=0.5,linestyle=':')
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green',lw=3, alpha=0.5,ls='dashed')
```
