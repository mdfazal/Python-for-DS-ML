# Matplotlib Concept Lecture 

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')

```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[1].plot(x,y)
axes[1].set_title('Second Plot')
axes[0].plot(y,x)
axes[0].set_title('First Plot')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X SQUARED')
ax.plot(x,x**3, label='X CUBED')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X SQUARED')
ax.plot(x,x**3, label='X CUBED')

ax.legend(loc=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X SQUARED')
ax.plot(x,x**3, label='X CUBED')

ax.legend(loc=(0.1,0.1))
```

# Exercise

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y, color='red')
ax2.plot(x,y, color='red')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

#Large
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

#Insert
ax2.plot(x,y)
ax2.set_title('Zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,axes= plt.subplots(1,2)
```

```python
fig,axes= plt.subplots(1,2)

axes[0].plot(x,y, color='blue', lw=3,ls='--')
axes[1].plot(x,z,color='red',lw=3)
```

```python
fig,axes= plt.subplots(1,2, figsize=(12,2))

axes[0].plot(x,y, color='blue', lw=3,ls='--')
axes[1].plot(x,z,color='red',lw=3)
```
